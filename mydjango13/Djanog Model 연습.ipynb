## Conditions

```python
qs = Shop.objects.all()
```

```python
# filter <-> exclude
qs.filter(pk=1)
```

```python
qs.exclude(pk=1)
```

```python
# 정수형 필드에 적용
# lt : less than            <
# lte : less than equal     <=
# gt : greater than         >
# gte : greater than equal  >=

qs.filter(
    pk__gte=5, # greater than equal
    pk__lt=10, # less than
)
```

```python
qs.filter(pk__gte=5).filter(pk__lt=10) # => 원본 qs 변화 X -> chaining
```

```python
qs1 = Shop.objects.all()
qs1 = qs1.filter(pk__gte=5)
qs1 = qs1.filter(pk__lt=10)
```

```python
qs2 = Shop.objects.all().filter(pk__gte=5).filter(pk__lt=10)
```

## Get 1 Row

```python
pk = 10
not_existed_pk = 100000
```

```python
shop = Shop.objects.get(pk=pk) # 가져올 대상이 1개임을 기대, 1개가 아니면 예외가 발생.
shop
```

```python
Shop.objects.get(pk=not_existed_pk) # Shop.DoesNotExist 예외가 발생
```

```python
Shop.objects.filter(pk=pk)[0] # 여러개임을 가정 -> 쿼리 셋 반환
shop
```

```python
Shop.objects.filter(pk=not_existed_pk)[0] # IndexError 예외가 발생
```

```python
Shop.objects.filter(pk=not_existed_pk).first() # first (<-> last): 지정 Row가 없다면 None을 반환할 뿐, 예외는 발생 X
```

## Conditinos (OR)

```python
qs = Shop.objects.all()
qs.filter(pk__gte=5, pk__lt=10) # AND 조건
```

```python
from django.db.models import Q
```

```python
Q(pk__gte=5)
```

```python
Q(pk__lt=10)
```

```python
Q(pk__gte=5) & Q(pk__lt=10)
```

```python
Q(pk__gte=5) | Q(pk__lt=10)
```

```python
qs = Shop.objects.all()
qs.filter(
    Q(pk__gte=5) | Q(pk__lt=10)
)
```

```python
conditions = Q(pk__gte=5)
# conditions = conditions | Q(pk__lt=10)
conditions |= Q(pk__lt=10)

conditions
```

## 지정 범위의 Rows 수정

```python
# 수행 되는 쿼리 1개 -> 더 효율적
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
qs.update(
    description="not used description",
)
```

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
for shop in qs:
    shop.description = "not used description"
    shop.save() # 되도록이면 루프문 안에서 하지 말고 쿼리 셋을 통해서
```
